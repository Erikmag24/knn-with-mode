### Learning from data

Come possiamo ora implementare un sistema capace di predirre la categoria di nuovi dati in base alle osservazioni precedenti?

Anzitutto, dobbiamo separare i dati disponibili in due gruppi, uno lo useremo per imparare (training set), l'altro (testing set) per verificare se il nostro modello funziona con dati nuovi.

Bisogna assicurarsi che tutte le classi siano equamente rappresentate nei due sets.

```python
X = iris.data
y = iris.target

# taking the first 60% of each class for training (90 points), and the rest 40% for testing (60 points)
X_train = np.concatenate((X[:30,:], X[50:80,:], X[100:130,:]))
X_test = np.concatenate((X[30:50,:], X[80:100,:], X[130:,:]))
y_train = np.concatenate((y[:30], y[50:80], y[100:130]))
y_test = np.concatenate((y[30:50], y[80:100], y[130:]))

print(np.shape(X_train), np.shape(X_test), len(y_train), len(y_test))

print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

```python
from scipy import stats
class knn:
    def __init__(self,x, x2, y2, knumber):
        self.x=x
        self.x2=x2
        self.y2=y2
        self.knumber=knumber
    def calculate(self):
        for element in self.x:
            distances=[]
            for element2 in self.x2:
                distanzapunto=0
                for elementx,elementx2 in zip(element,element2):
                    distanzapunto+=(elementx-elementx2)**2
                distances.append(distanzapunto**1/2)
                print(distanzapunto**1/2)
            mode=[]
            modes=(np.argsort(distances)[:self.knumber])
            for distance in modes:
                mode.append(self.y2[distance])
            print(stats.mode(np.array(mode)))

knn(X_test, X_train, y_train,1).calculate()
```

```python
from statistics import mode
class KNeighborsClassifier:
    def __init__(self, n_neighbors=5):
        self.n_neighbors=n_neighbors
        
        
        
        
    def fit(self,X, y):
        self.X=X
        self.y=y
        
        
        
    def predict(self, X):
        self.x2=X
        self.modes=[]
        self.numero=0
        for element2 in self.x2:
            self.numero+=1
            distances=[]
            for element in self.X:
                distanzapunto=0
                for elementx,elementx2 in zip(element,element2):
                    distanzapunto+=(elementx-elementx2)**2
                distances.append(distanzapunto**(1/2))
            modex=[]  
            distances=(np.argsort(distances)[:self.n_neighbors])
            for distance in distances:
                modex.append(self.y[distance])
                
            self.modes.append(mode(modex))
            #print(self.modes)
        
        
    def errors(self,y):
        errors=0
        self.y2=y
        for X,y in  zip(self.modes, self.y2):
            if X!=y:
                errors+=1
        return(100-(errors*100)/self.numero)
        

        
listaerrori=[]
for i in range(1,100):
    knn=KNeighborsClassifier(i)
    knn.fit(X_train,y_train)
    knn.predict(X_test)
    listaerrori.append(knn.errors(y_test))
```

```python
import matplotlib.pyplot as plt
plt.plot(listaerrori)
```
